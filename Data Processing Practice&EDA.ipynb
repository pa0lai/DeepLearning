```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

Dataset_File = "WeatherData.csv"
if not os.path.isfile(Dataset_File):
  os.system("wget https://raw.githubusercontent.com/cnchi/datasets/master/" + Dataset_File)

```

```python
dfCSV = pd.read_csv("WeatherData.csv")


print(dfCSV) #題目要求使用print()
dfCSV.head() #參考輸出結果
```

```python
print("平均溫度: " + dfCSV["氣溫(℃)"].mean().astype(str) + " ℃")
print("最高濕度: " + dfCSV["相對溼度(%)"].max().astype(str) + " %")
print("最低氣壓: " + dfCSV["當日氣壓(hPa)"].min().astype(str) + " hPa")
print("總雨量: " + dfCSV["降水量(mm)"].sum().astype(str) + " mm")
```

繪圖

```python
X = np.arange(0., 30., 1)

plt.title("2023年6月溫度折線圖", fontproperties=myfont)
plt.xlabel("日期", fontproperties=myfont)
plt.ylabel("溫度(℃)", fontproperties=myfont)

plt.plot(X, dfCSV.loc[X, ["氣溫(℃)"]] ,'o-', label="溫度")
plt.legend(loc = 'best', prop=myfont)

# plt.xkcd()
plt.show()
```
